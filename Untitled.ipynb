```python
def dict_to_list(dict1):
    genre_list=[]
    for i in ast.literal_eval(dict1):
        genre_list.append(i['name'])
    return genre_list
```

```python
def for_cast(cast_name):
    list_for_cast=[]
    counter=0
    for i in ast.literal_eval(cast_name):
        if counter!=3:
            list_for_cast.append(i['name'])
            counter+=1
        else:
            break
    return list_for_cast
```

```python
def for_directors(obj):
    dir_list=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            dir_list.append(i['name'])
    return dir_list        
    
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
new_dataset['tags'][0]
```

```python
def stemmer(text):
    st=[]
    for i in text.split():
        st.append(ps.stem(i))
    return " ".join(st)
```
